```python
dataset["Location_Type"]=dataset["Location_Type"].map({"L1":1,"L2":2,"L3":3,"L4":4,"L5":5})
```

```python
dataset
```

```python
dataset["Region_Code"]=dataset["Region_Code"].map({"R1":1,"R2":2,"R3":3,"R4":4,"R5":5})
```

```python
dataset
```

```python
dataset["Discount"]=dataset["Discount"].map({"Yes":1,"No":0})
```

```python
dataset
```

```python
x= np.array(dataset[["Store_Type","Location_Type","Region_Code","Discount"]])
y= np.array(dataset["#Order"])
```

```python
x
```

```python
y
```

```python
# building ml model
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test  = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
len(x_train)
```

```python
#large dataset so pip install ligthgbm
import lightgbm as ltb
model = ltb.LGBMRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred= model.predict(x_test)
```
